```python
#Handle Skewed Distributions for 'close_value', 'sales_price', 'revenue' and 'employees' with Log Transformation
df['log_close_value'] = np.log1p(df['close_value'])  # log1p is log(1 + x)
df['log_sales_price'] = np.log1p(df['sales_price'])
df['log_revenue'] = np.log1p(df['revenue'])
df['log_employees'] = np.log1p(df['employees'])
```

```python
#Handle Skewed Distributions for 'close_value', 'sales_price' and 'revenue' with Box-Cox Transformation
df['boxcox_close_value'], _ = stats.boxcox(df['close_value'] + 1)  # Adding 1 to handle zero values
df['boxcox_sales_price'], _ = stats.boxcox(df['sales_price'] + 1)
df['boxcox_revenue'], _ = stats.boxcox(df['revenue'] + 1)
df['boxcox_employees'], _ = stats.boxcox(df['employees'] + 1)
```

```python
# Handle Outliers: Using IQR
def iqr_filter(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series[(series >= lower_bound) & (series <= upper_bound)]

df['iqr_close_value'] = iqr_filter(df['close_value'])
df['iqr_sales_price'] = iqr_filter(df['sales_price'])
df['iqr_revenue'] = iqr_filter(df['revenue'])
df['iqr_employees'] = iqr_filter(df['employees'])
```

# **5. Model Training and Evaluation**

## 5.1 Linear Regresion
